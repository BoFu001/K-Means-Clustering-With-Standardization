## K-Means Clustering: With Standardization and Inverse Transform

This notebook demonstrates K-Means clustering **after standardizing** the features using `StandardScaler`. It also performs **inverse transformation** of cluster centers for visualization in the original space. I'll explore:
- Data loading and visualization
- Choosing the optimal number of clusters using the Elbow Method
- Applying K-Means and visualizing the result

```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Step 2: Load and Preview the Data
df = pd.read_csv("data.csv")
df.head()
```

```python
# Step 3: Standardize the Features
# Standardization helps K-Means perform better when variables are on different scales.
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['V1', 'V2']])
```

```python
# Step 4: Determine Optimal Number of Clusters (K) using Elbow Method
wcss = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=15)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(K_range, wcss, 'bo-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method For Optimal K (Scaled Data)')
plt.show()
```

```python
# Step 5: Apply KMeans with Optimal K
# Replace with chosen K from the elbow plot
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=15)
clusters = kmeans.fit_predict(df_scaled)

# Add the cluster labels to the original DataFrame
df['Cluster'] = clusters
```

```python
# Step 6: Inverse Transform the Cluster Centers to Original Space
centers_original = scaler.inverse_transform(kmeans.cluster_centers_)
```

```python
# Step 7: Visualize Clusters in Original Feature Space
plt.scatter(df.V1, df.V2, c=df.Cluster, cmap='viridis', alpha=0.3, s=15)
plt.scatter(centers_original[:, 0], centers_original[:, 1], s=60, c='red', marker='X', label='Centroids (Original Scale)')
plt.xlabel('V1')
plt.ylabel('V2')
plt.title(f'KMeans Clustering with K={optimal_k} (Standardized + Inverse Transformed)')
plt.legend()
plt.show()
```

## Conclusion
- Features were standardized before clustering, ensuring equal importance.
- We used the elbow method to determine the best number of clusters.
- Cluster centers were transformed back to the original scale for accurate visualization.
